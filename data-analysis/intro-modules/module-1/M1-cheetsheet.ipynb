## Matplotlib


```python
import random as r
import matplotlib.pyplot as plt

from typing import  List
```

```python
#  utils

def generate_array(size:int)->List[int]:
    # initialize a 1D array with size and default value
    arry = [None] * size

    # populate random data
    for idx in range(0, len(arry)):
        arry[idx] = r.randint(0,50)

    return arry
```

```python
# 1D plot

arry = generate_array(30)
plt.plot(arry)
```

```python
#  2D plot

arry_x = generate_array(15)
arry_y = generate_array(15)

plt.plot(arry_x, arry_y)
plt.xlabel("X-axis")
plt.ylabel("y-axis")
plt.title("Random 2D plot")
plt.legend(['data'])
plt.show()
```

```python
# multiple datasets on same graph

# line chart

arry_new = generate_array(15)

plt.figure(figsize=(9,5)) 
plt.plot(arry_x, arry_y, marker="o", color="brown",ls="--")
plt.plot(arry_x, arry_new, marker="x", color="teal", ls="-.")

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Random Line plot with two datasets")
plt.legend(["Old data", "New Data"])
```

```python
# scatter plot

plt.figure(figsize=(9,5))
plt.scatter(arry_x, arry_y, color="lime")
plt.scatter(arry_x, arry_new, color="purple")

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Random Line plot with two datasets")
plt.legend(["Old data", "New Data"])
```

```python
# bar chart

latest_arry = generate_array(15)

plt.bar(arry_x, arry_y, color="purple")
plt.bar(arry_x, arry_new, bottom=arry_y, color="pink")
plt.bar(arry_x, latest_arry, bottom=arry_y, color="lime")

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Random Line plot with two datasets")
plt.legend(["Old data", "New Data", "latest data"])
plt.savefig("rand.png")
```
