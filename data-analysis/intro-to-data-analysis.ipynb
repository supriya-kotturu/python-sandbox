```python
# 9. Find indices of non-zero elements from [1,2,0,0,4,0]

def get_valid_idx (arr: List[int]) -> List[int]:
    idx_list = []

    for idx, val in enumerate(arr):
        if val != 0:
            idx_list.append(idx)

    return idx_list

get_valid_idx([1,2,0,0,4,0])
```

```python
# 14. How to get the positions where elements of two arrays match?

def get_positions(arr_1: List[int], arr_2: List[int]) -> List[int]:
    pos_map = {}

    for idx, ele in enumerate(arr_1):
        pos_map[ele]=[idx]

    for idx, ele in enumerate(arr_2):
        if(ele in pos_map):
            pos_map[ele].append(idx)

    for ele in list(pos_map.keys()):
        if(len(pos_map[ele]) <2):
            del pos_map[ele]

    return pos_map

arr_1 = np.array(range(1,45,2))
arr_2 = np.array(range(1,45,4))

get_positions(arr_1, arr_2)
```

```python
# 16. How to compute the min-by-max for each row for a numpy array 2d?

matrix = horizontal[:-1].reshape(3,3)
min_by_max = np.zeros(matrix.shape[0])

for idx, row in enumerate(matrix):
    min_val = min(row)
    max_val = max(row)
    min_by_max[idx] = min_val /max_val

print(min_by_max)

```

```python
# 18. What are the purpose of the following statements:
#   Df.column
#   Df.iloc[:,:-5]
#   Df[2:8]
#   Df[:]
#   Df.iloc[:-4,:]

df = pd.DataFrame({"height": height, "weight": weight})
df.height           # returns the column values

df.iloc[:, :-1]     # used for integer-location based indexing
# : before the comma indicates selecting all rows.
# :-1 after the comma indicates selecting all columns except for the last 1. 
# It is useful for dropping the last few columns from the DataFrame.

df[2:3]             # returns a subset of the dataframe
df[:]               # returns all the values in the dataframe

df.iloc[:-4,:]        # used for integer-location based indexing
# :-4 before the comma indicates selecting all columns except for the last 4. 
# : after the comma indicates selecting all columns.
# It is useful for dropping the last few rows from the DataFrame.
```

```python
# 26. build a sales dataframe and draw the following insights
#   a. find the total sales per state
#   b. find the total sales per employee
#   c. find the average sales on both employee and state 
#   d. find mean, median and min sale state wise

employees = ['Sahay','George','Priya','Manila','Raina','Manila','Priya']
sales = [125600,235600,213400,189000,456000,172000,201400]
quarters = [1,1,1,1,1,2,2]
states = ['Delhi','Tamil Nadu','Kerala','Haryana','West Bengal','Haryana','Kerala']

sales_df = pd.DataFrame({"Employee": employees, "Sales":sales, "Quarter":quarters,'State': states})

#   a. find the total sales per state
sales_per_state = sales_df.groupby('State')['Sales'].sum()
print(sales_per_state, end="\n\n")

#   b. find the total sales per employee
sales_per_employee = sales_df.groupby('Employee')['Sales'].sum()
print(sales_per_employee, end="\n\n")

#   c. find the average sales on both employee and state 
avg_sales_per_employee = sales_df.groupby('Employee')['Sales'].mean()
print(avg_sales_per_employee, end="\n\n")

avg_sales_per_state = sales_df.groupby("State")['Sales'].mean()
print (avg_sales_per_state, end="\n\n")

#   d. find mean, median and min sale state wise
mean_sales = sales_df['Sales'].mean()
median_sales = sales_df['Sales'].median()
min_sales = sales_df['Sales'].min()

print("Mean sales:{}, Median sales:{} Min sales:{}".format(mean_sales, median_sales, min_sales))

```
