# K-means

```python
import pandas as pd  # Importing pandas for data manipulation
from sklearn.datasets import load_breast_cancer  # To load the Wisconsin Breast Cancer dataset
from sklearn.cluster import KMeans  # To use K-Means clustering
import matplotlib.pyplot as plt  # For plotting
from sklearn.decomposition import PCA  # To reduce dimensions for visualization
```

```python
# Load the Breast Cancer dataset from sklearn
file_path = "./wisconsin-breast-cancer.csv"  # Update with your actual CSV file path
df = pd.read_csv(file_path)

# Inspect the first few rows to understand the structure
print(df.head())
```

```python
# Drop unnecessary columns (like 'id' or 'Unnamed: 32')
df = df.drop(columns=['id', 'Unnamed: 32'], errors='ignore')
```

```python
# Separate features and target
X = df.drop(columns=['diagnosis'])  # Feature matrix
y = df['diagnosis']  # Target variable (not used in clustering)
```

```python
# Initialize the K-Means model
kmeans = KMeans(n_clusters=2, random_state=42)
```

```python
# Fit the model to the data
kmeans.fit(X)
```

```python
# Get cluster labels
labels = kmeans.labels_
```

```python
# Reduce dimensions for visualization using PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
# Create a scatter plot of the two clusters
plt.figure(figsize=(10, 8))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='Accent', edgecolor='k', s=50)
plt.title('K-Means Clustering on Wisconsin Breast Cancer Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
# Print cluster centers
print("Cluster centers:")
print(kmeans.cluster_centers_)
```
