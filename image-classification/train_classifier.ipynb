# Train Classifier

```python
import pickle
import numpy as np

from sklearn.ensemble import  RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
```

```python
data_dict = pickle.load(open("./data.pickle", "rb"))

print(data_dict.keys())
print((data_dict['data']))
```

```python
# Access the data and labels using the correct keys
data_array = data_dict["data"]
labels = data_dict["labels"]

# Verify the shape of each element in the data list
for i, element in enumerate(data_array):
    print(f"Element {i} shape: {np.shape(element)}")

# Ensure all elements are of the same length and type
# Assuming each element should be a 1D array of the same length
filtered_data = []
filtered_labels = []
expected_length = len(data_array[0])

for element, label in zip(data_array, labels):
    if isinstance(element, list) and len(element) == expected_length:
        filtered_data.append(element)
        filtered_labels.append(label)

# Convert the filtered data and labels to NumPy arrays
data = np.array(filtered_data)
labels = np.array(filtered_labels)

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(data, labels, test_size=0.2, shuffle=True, stratify=labels, random_state=1278)
```

```python
model = RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
y_predict = model.predict(x_test)
```

```python
score = accuracy_score(y_predict, y_test)
```

```python
print(f"{score* 100}")
```

```python
# Calculate evaluation metrics
precision = precision_score(y_test, y_predict, average='weighted')  # 'weighted' for multi-class
recall = recall_score(y_test, y_predict, average='weighted')
f1 = f1_score(y_test, y_predict, average='weighted')

print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-score: {f1:.2f}")
```

```python
f = open('model.p', 'wb')
pickle.dump({'model': model}, f)
f.close()
```
