```python
# Bank Account to withdraw and deposit amount

class BankAccount:
    def __init__(self,account_number: str, account_holder_name: str, balance: int) -> None:
        self.account_number = account_number
        self.account_holder_name = account_holder_name
        self.__balance = balance
        
    def deposit(self, deposite_amount: int) -> None:
        self.__balance += deposite_amount
        self.getUpdatedBalance()
        
    def withdraw(self, withdraw_amount: int) -> None:
        self.__balance -= withdraw_amount
        self.getUpdatedBalance()
        
    def getUpdatedBalance(self)->None:
        print(f"Updated Balance: {self.__balance}")
    
    def getAccountDetails(self)->None:
        print(f"Account Holder Name: {self.account_holder_name}")
        print(f"Account Number: {self.account_number}")
        print(f"Balance: {self.__balance}\n------\n")
        
        
        
holder = BankAccount(account_number="KKBK034011167", account_holder_name="Rita", balance=238000)
holder.getAccountDetails()
holder.deposit(1000)
holder.withdraw(5000)
        
```

```python
# Employee Management to calculate yearly bonus

class Salary:
    def __init__(self, basic:int) -> None:
        self._basic_amount = basic
        self._hra_amount = 0.3 * basic
        self._special_allowance = 0.1 * basic
        self._leave_travel_allowance = 0.05 * basic
        self.calculateGrossSalary()
        
    def calculateGrossSalary(self)-> None :
        self._gross_salary =  self._basic_amount + self._hra_amount + self._special_allowance + self._leave_travel_allowance
    
    def grossSalary(self) -> int: return self._gross_salary
    
    def basic(self) -> int: return self._basic_amount
    
    def hra(self) -> int: return self._hra_amount
    
    def specialAllowance(self) -> int: return self._special_allowance
    
    def leaveTravelAllowance(self) -> int: return self._leave_travel_allowance



class Employee:
    def __init__(self,id: int, name:str, basic_salary_amount: int) -> None:
        self._id = id
        self._name = name
        self._salary = Salary(basic_salary_amount)
        
    def getName(self)->str: return self._name
    
    def getSalary(self)->int: return self._salary
    
    def getId(self)->int: return self._id
    
    def getEmployeeDetails(self)-> None:
        print(f"Name: {self._name}")
        print(f"ID: {self._id}")
        print(f"Salary: {self._salary}")
    
    def calculateYearlyBonus(self)->int:
        # 1. 10% - basic salary <= 15,000
        # 2. 8% - basic salary > 15,000 and <= 20,000
        # 3. 7% - basic salary > 20,000
        
        basicSalary = self._salary.basic()
        bonus = 0
        
        if (basicSalary <= 15000): bonus += basicSalary * 0.1
        elif (basicSalary > 15000 and basicSalary <= 20000) : bonus += basicSalary * 0.08
        else: bonus += basicSalary * 0.07
        
        return bonus
        
        
emp = Employee(id=784512, name="Priyanka", basic_salary_amount=34000)
empBonus = emp.calculateYearlyBonus()
print(f"Expected Yearly Bonus: {empBonus}")
        
```

```python
# Vehicle Rental System to rent, return, and display available vehicles

class Vehicle:
    def __init__(self, vehicleIdentificationNumber: str, brand: str, model: str, color:str,  mileage: int, fuelType: str, kilometersTravelled:int) -> None:
        self.__vin = vehicleIdentificationNumber
        self.__fuelType = fuelType
        self.__brand = brand
        self.__model = model
        self.__color = color
        self.__mileage = mileage
        self.__kilometersTravelled = kilometersTravelled
        
    def vehicleIdentificationNumber(self) -> str: return self.__vin
    
    def brand(self): return self.__brand
    
    def model(self) -> str: return self.__model
    
    def color(self) -> str: return self.__color
    
    def fuelType(self) -> str: return self.__fuelType
    
    def showDetails(self) -> None:
        print(f"VIN: {self.__vin}")
        print(f"Fuel Type: {self.fuelType()}")
        print(f"Brand: {self.brand()}")
        print(f"Model: {self.model()}")
        print(f"Color: {self.color()}")
        print(f"Mileage: {self.__mileage}")
        print(f"Km Travelled: {self.__kilometersTravelled}")
        print(f"-------\n")
        
    def updateMileage(self, mileage) -> None:
        self.__mileage += mileage
        
    def updateKilometersTravelled(self, km) -> None:
        if(self.__kilometersTravelled > km): 
            print(f"Kilometers travelled cannot be decreased.")
            return
        self.__kilometersTravelled = km
        
        
class VehicleRentalSystem:
    def __init__(self, available: List[Vehicle], rented: List[Vehicle] = []):
        self.available = available
        self.rented = rented
        
    def showAvailableVehicles(self) -> None:
        print(f"AVAILABLE VEHICLES")
        for vehicle in self.available:
            vehicle.showDetails()
            
    def addVehicle(self, vehicle: Vehicle) -> None:
        self.available.append(vehicle)
        print(f"Available vehicle list updated.")
        
    def removeVehicle(self, vehicleIdentificationNumber: str) -> None:
        for idx, vehicle in enumerate(self.available):
            if vehicle.vehicleIdentificationNumber() == vehicleIdentificationNumber:
                del self.available[idx]
                return
                
    def rentVehicle(self, vehicleIdentificationNumber: str) -> None:
        for idx, vehicle in enumerate(self.available):
            if vehicle.vehicleIdentificationNumber() == vehicleIdentificationNumber:
                rentedVehicle = (vehicle)
                
                self.rented.append(rentedVehicle)
                print(f"You have rented vehicle with VIN: {rentedVehicle.vehicleIdentificationNumber()}")
                
                del self.available[idx]
                return
            
        print(f"Vehicle VIN: {vehicleIdentificationNumber} is currently unavailable or rented out.")
                
        
    def returnVehicle(self, vehicleIdentificationNumber: str, kilometersTravelled) -> None:
        for idx, vehicle in enumerate(self.rented):
            if vehicle.vehicleIdentificationNumber() == vehicleIdentificationNumber:
                returnedVehicle = (vehicle)
                returnedVehicle.updateKilometersTravelled(kilometersTravelled)
                
                self.available.append(returnedVehicle)
                print(f"You have returned vehicle with VIN: {returnedVehicle.vehicleIdentificationNumber()}")
            
                del self.rented[idx]
                return
        
        print(f"Vehicle VIN: {vehicleIdentificationNumber} is not rented out earlier.")
        
    
```

```python
# 4. Library Catalog

class Book:
    def __init__(self, id: int, title: str, author: str, rentedBy: str = None, isAvailable:bool = True):
        self.__id = id
        self.__title = title
        self.__author = author
        self.__rentedBy = rentedBy
        self.__isAvailable = isAvailable
        
    def id(self) -> int: return self.__id
    
    def title(self) -> str: return self.__title
    
    def author(self) -> str: return self.__author
    
    def rentedBy(self) -> str: return self.__rentedBy
    
    def isAvailable(self) -> bool: return self.__isAvailable
    
    def isRentedBy(self, name: str) : 
        self.__rentedBy = name
        self.__isAvailable = False
        
    def isReturned(self):
        self.__isAvailable = True
        self.__rentedBy = None
        
    def getDetails(self): 
        print(f"ID: {self.id()}")
        print(f"Title: {self.title()}")
        print(f"Author: {self.author()}")
        if not self.isAvailable(): 
            print(f"Rented by: {self.rentedBy()}")
        print("------")
        
        
class Library:
    def __init__(self, books: List[Book]) -> None:
        self.__books = books
        
    def __bookMap(self, func: Callable, books: List[Book]) -> List[Book]:
        length =len(books)
        result = []
        
        for idx in range(length):
            if (func(books[idx])):
                result.append(books[idx])
            
        return result
    
    def addBook(self, book: Book) -> None:
        self.__books.append(book)
        print(f"Available book list updated.")
        
        
    def getAvailableBooks(self):
        isAvailable = lambda book: book.isAvailable()
        return self.__bookMap(isAvailable, self.__books)
    
    def getRentedBooks(self):
        isRented  = lambda book: not book.isAvailable()
        return self.__bookMap(isRented, self.__books)
    
    def rentBook(self, bookId: int, name: str) -> None:
        for book in self.__books:
            if book.id() == bookId and book.isAvailable(): 
                book.isRentedBy(name)
                print(f"Book '{book.title()}' has been rented by {name}.")
                return
        print(f"Book not available for rent.")
    
    def returnBook(self, bookId: int):
        for book in self.__books:
            if book.id() == bookId and not book.isAvailable():
                book.isReturned()
                print(f"Book {book.title()} has been returned.")
                return
        print(f"Book was not rented or does not exist.")
            
    def getAllBooksdetails(self): 
        for book in self.__books:
            book.getDetails()
            
    def getAvailableBookDetails(self):
        print(f"\nAVAILABLE BOOKS")
        availableBooks = self.getAvailableBooks()
        
        for book in availableBooks:
            book.getDetails()
            
    def getRentedBookDetails(self):
        print(f"\nRENTED BOOKS")
        rentedBooks = self.getRentedBooks()
        
        for book in rentedBooks:
            book.getDetails()
```

```python
#  5. Student Management System to calculate average grade

class Student:
    def __init__(self, id: int, name: str, grades: List[int]):
        self.__id = id
        self.__name = name
        self.__grades = grades
        
    def studentId(self): return self.__id
    
    def name(self): return self.__name
    
    def grades(self): return self.__grades
    
    def getAverageGrade(self): return sum(self.__grades) // len(self.__grades)
    
    def showDetails(self): 
        print(f"Id: {self.studentId()}")
        print(f"Name: {self.name()}")
        print(f"Grades: {self.grades()}")
        print(f"Average Grade: {self.getAverageGrade()}")
            
            
std = Student(101, "Aashish", [78,89,67,89,78,90])
std.showDetails()
        
```

```python
# 6. Social Media Profile Creator

class Post:
    def __init__(self, id: str, title: str, description: str):
        self.__id = id 
        self.__title = title
        self.__description = description
    
    @property
    def postId(self): return self.__id
    
    @property
    def title(self): return self.__title
    
    @property
    def description(self): return self.__description
    
    def getPostDetails(self):
        print(f"\nId : {self.postId}")
        print(f"Title: {self.title}")
        print(f"Description: {self.description}")
        

class SocialMediaProfile: 
    def __init__(self, id: int, name: str, contact: str, gender: str):
        self.__id = id
        self.__name = name
        self.__contact = contact
        self.__gender = gender
        self.__posts: List[Post] = []
        
    def profileId (self): return self.__id
    
    def name(self): return self.__name
    
    def contact(self): return self.__contact
    
    def gender(self): return self.__gender
    
    def getPosts(self): return self.__posts
    
    def addPost(self, post: Post):
        self.__posts.append(post)
        
    def deletePost(self, postId: str):
        for idx, post in enumerate(self.__posts):
            if postId == post.postId:
                print("IDx: ", idx)
                del self.__posts[idx]
                return
            
        print(f"No Post found with ID: {postId}")
        
    def showPosts(self):
        if(len(self.__posts) == 0):
            print("No Posts.")
            return
        
        for post in self.__posts:
            post.getPostDetails()
                
        
smp = SocialMediaProfile(121, "Leon", "7789654122", 'male')

post1 = Post(id="1", title="Hello World", description="This is my first post.")
post2 = Post(id="2", title="Python Tips", description="Some useful tips for Python programming.")
post3 = Post(id="3", title="Travel Diaries", description="Sharing my travel experiences from around the world.")

# smp.showPosts()

smp.addPost(post1)
smp.addPost(post2)
smp.addPost(post3)

# smp.showPosts()

smp.deletePost('2')
smp.showPosts()
```
